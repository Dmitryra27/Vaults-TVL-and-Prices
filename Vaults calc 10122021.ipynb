```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load data
data = pd.read_csv('vaultsTvl101221.csv')
```

```python
fig = make_subplots(subplot_titles=('' ), specs=[[{"secondary_y": True}]])
#x = range(0,len(data['Date']))
x= np.arange(0, 185, 1)

fig.add_trace(go.Scatter(x=x, y=data['MIM Price'], line=dict(color='red'), name='MIM Price'), secondary_y=False)
fig.add_trace(go.Scatter(x=x, y=data['EURS Price'], line=dict(color='green'), name='EURS Price'), secondary_y=False)
fig.add_trace(go.Scatter(x=x, y=data['FRAX Price'], line=dict(color='blue'), name='FRAX Price'), secondary_y=False)
fig.add_trace(go.Scatter(x=x, y=data['USDN Price'], line=dict(color='yellow'), name='USDN Price'), secondary_y=False)
#fig.add_trace(go.Scatter(x=x, y=data['mUSD Price'], line=dict(color='black'), name='mUSD Price'), secondary_y=False)
fig.add_trace(go.Scatter(x=x, y=data['USDP Price'], line=dict(color='brown'), name='USDP Price'), secondary_y=False)
fig.add_trace(go.Scatter(x=x, y=data['IronBank Price'], line=dict(color='grey'), name='IronBank Price'), secondary_y=False)
fig.add_trace(go.Scatter(x=x, y=data['sUSD Price'], line=dict(color='silver'), name='sUSD Price'), secondary_y=False)
#fig.add_trace(go.Scatter(x=x, y=data['alUSD Price'], line=dict(color='darkgreen'), name='alUSD Price'), secondary_y=False)
fig.update_layout(title_text='<b>Choosen by Price Rase  Vaults</b>')
fig.update_xaxes(title_text='Step')
fig.update_layout(yaxis_range=[1,1.15])

fig.show()
```
